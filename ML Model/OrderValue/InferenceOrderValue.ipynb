```python
import joblib
import numpy as np
import pandas as pd

model = joblib.load("final_spend_model_xgb.pkl")
training_columns = joblib.load("xgb_training_columns.pkl")
```

```python
def predict_spend(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    
    df = df.drop(columns=['target_avg_order_value'], errors='ignore')

    
    X_new = df.reindex(columns=training_columns)

    
    preds_log = model.predict(X_new)
    preds_dollars = np.expm1(preds_log)

    
    result = pd.DataFrame({
        "customer_unique_id": df["customer_unique_id"],
        "predicted_avg_order_value": preds_dollars
    })
    return result
```

```python
master_df=pd.read_csv("../Processed Data/prd_order_value.csv")
scored_df = predict_spend(master_df)
scored_df.to_csv("repeat_customers_scored.csv", index=False)
```
