```python
import pandas as pd 
df_master=pd.read_csv("../Processed Data/prd_master_orders_customers.csv")
```

```python
import pandas as pd


df_master['order_purchase_timestamp'] = pd.to_datetime(df_master['order_purchase_timestamp'])

total_counts = df_master.groupby('customer_unique_id')['order_id'].nunique().reset_index()
total_counts.columns = ['customer_unique_id', 'total_lifetime_orders']


df_first_orders = df_master[df_master['is_first_order'] == 1].copy()


df_loyal = df_first_orders.merge(total_counts, on='customer_unique_id', how='left')

df_loyal['label_is_loyal'] = (df_loyal['total_lifetime_orders'] > 1).astype(int)

df_loyal = df_loyal.drop(columns=['total_lifetime_orders'])
```

```python
df_loyal['label_is_loyal'].value_counts()
```

```python
df_loyal.to_csv("../Processed Data/prd_loyal_customers.csv", index=False)
```
